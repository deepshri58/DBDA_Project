# Decompose TS & Vizualize Components of TS

# Check the Stationarity

## ADF Augmented Dickey Fuller Test

```python
from statsmodels.tsa.stattools import adfuller
```

```python
adfuller_result = adfuller(data['Close'])
print('ADF Statistic: ', adfuller_result[0])
print('p-value: ', adfuller_result[1])
```

- If p value is greater than 0.05 then that data is not stationary

#data is not stationary thats why we are going to make the data as stationary

```python
# Diff Method
data_diff = data.diff().dropna()
diff_adfuller_result = adfuller(data_diff['Close'])
print('For difference Method ADF Statistic is: ', diff_adfuller_result[0])
print('For difference Method p-value is: ', diff_adfuller_result[1])
```

```python
#sqrt method
data_sqrt = np.sqrt(data).dropna()
sqrt_adfuller_result = adfuller(data_sqrt['Close'])
print('For square Root Method ADF Statistic is: ', sqrt_adfuller_result[0])
print('For square Root Method p-value is: ', sqrt_adfuller_result[1])
```

```python
# Diff Twice Method
data_difftwice = data.diff().diff().dropna()
difftwice_adfuller_result = adfuller(data_difftwice['Close'])
print('For difference Method ADF Statistic: ', difftwice_adfuller_result[0])
print('Fordifference Method p-value: ', difftwice_adfuller_result[1])
```

```python
# Plot the time series before doing data as stationary
plt.figure(figsize=(16,8))
sns.lineplot(x=data.index, y='Close', data=data, linewidth=1.5, label='Before Stationary').set_title('Time Series data Before stationary')
plt.show()

# Plot the time series after stationary
plt.figure(figsize=(16,8))
sns.lineplot(x=data_difftwice.index, y='Close', data=data_difftwice, label='After Stationary', color='green').set_title('Stationary Time Series as Difftwice method')
plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
# Plot ACF and PACF with stationary data_difftwice

fig_diff, (ax1, ax2) = plt.subplots(2,1, figsize=(10,8))
# Plot ACF for data_diff
plot_acf(data_difftwice['Close'], lags=15, zero=False, ax=ax1, title='Autocorrelation as Differenced twice data')

# Plot PACF for data_difftwice
plot_pacf(data_difftwice['Close'], lags=15, zero=False, ax=ax2, title='Partial Autocorrelation as Differenced twice data')
plt.show()
```

```python
# Plot ACF and PACF for non-stationary data 

fig_data, (ax1, ax2) = plt.subplots(2,1, figsize=(10,8))
# Plot ACF of data
plot_acf(data['Close'], lags=10, zero=False, ax=ax1, title='Autocorrelation for Non-Stationary data')

# Plot PACF of data
plot_pacf(data['Close'], lags=10, zero=False, ax=ax2, title='Partial Autocorrelation for Non-Stationary data')
plt.show()
```

```python
arima_data = data.drop(columns=['High', 'Low', 'Open', 'Adj Close', 'Volume'])
arima_data.head()
```

```python
from pmdarima.arima import auto_arima
```

```python
results = auto_arima(arima_data,
                    seasonal=True,
                    start_p = 1,
                    start_q = 1,
                    start_P=1,
                    start_Q=1,
                    max_P=3,
                    max_Q=3,
                    m=7,                           # m is for seasonal period we are gonna take this as daily
                    information_criterion='aic',
                    trace=True,
                    error_action='ignore',
                    stepwise=True)
```

```python
# Based on this auto arima, order is (1, 1, 0) seasonal order=(0, 0, 0, 7)
# Spliting the data as train and test data 
auto_arima_train_data = arima_data.loc[:'2017']
auto_arima_test_data = arima_data.loc['2018':]

# Create a model as auto arima
auto_arima_model = SARIMAX(auto_arima_train_data,
                    seasonal=True,
                    order=(1,1,0),
                    seasonal_order=(0,0,0,7),
                    trend='c')
auto_arima_results = auto_arima_model.fit() 
auto_arima_pred_1year_traindata = auto_arima_results.get_prediction(start=-365, dynamic=False)

# predicting mean for 365 days
auto_arima_pred_mean_1year_traindata = auto_arima_pred_1year_traindata.predicted_mean

# Get confidence intervals of predicted
auto_arima_confidence_intervals = auto_arima_pred_1year_traindata.conf_int()

# Select lower and upper confidence limits
auto_arima_lower_limits = auto_arima_confidence_intervals.loc[:,'lower Close']
auto_arima_upper_limits = auto_arima_confidence_intervals.loc[:,'upper Close']

auto_arima_pred_mean_1year_traindata_df = auto_arima_pred_mean_1year_traindata.to_frame(name='pred_gold_mean')
auto_arima_pred_mean_1year_traindata_df.head()
```

```python
# Calculate MAE, MSE, RMSE
real_test_values = test_data['Close']

# Calculate MAE, MSE, RMSE
print('MAE: {}'.format(mean_absolute_error(real_test_values, auto_arima_mean_forecast)))
print('MSE: {}'.format(mean_squared_error(real_test_values, auto_arima_mean_forecast)))
print('RMSE: {}'.format(np.sqrt(mean_squared_error(real_test_values, auto_arima_mean_forecast))))
```
